```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


step_data = pd.read_csv("steps.csv", sep=";")
survey_data = pd.read_csv('survey.csv')
```

```python

df = survey_data.merge(step_data, on = "id")
df.head()
```

```python
df.describe()
```

Maak een boxplot van gewicht, en converteer dan de onmogelijke waarde naar een NaN.

```python
sns.boxplot(x="weight", data=df)
plt.title("Weight")
plt.xlabel("Weight")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
df['weight']=df['weight'].replace(700,np.nan)
```

Check with info() which column you want to calculate the mean over. Use the attribute verbose to display all columns

```python
df.info(verbose=True)
```

Calculate the mean over all columns for each row. Make sure to add axis=1

```python
df['steps_mean']=df.iloc[:,9:336].mean(axis=1)
df.head()
```

Create a histogram and boxplot of steps_mean, and calculate mean, median and std dev.

```python
sns.displot(x="steps_mean", kde=False, data=df) #kde=False means we don't get an overlying density plot
plt.title("Mean steps per day")
plt.xlabel("Mean steps")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x="steps_mean", data=df)
plt.title("Mean steps per day")
plt.xlabel("Mean steps")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
print('Median = '+str(df['steps_mean'].median()))
print('Mean = '+str(df['steps_mean'].mean()))
print('St Dev. = '+str(df['steps_mean'].std()))
```

you have clearly some outliers and a right skew, more than 30.000 steps would hypothetically be possible, but you can also see that these outlier influence the mean and standard deviation significantly, so I choose to filter out steps above 30.000.

```python
df=df[df['steps_mean']<30000]
```

```python
print('Median = '+str(df['steps_mean'].median()))
print('Mean = '+str(df['steps_mean'].mean()))
print('St Dev. = '+str(df['steps_mean'].std()))
```

```python
sns.displot(x="steps_mean", kde=False, data=df) #kde=False means we don't get an overlying density plot
plt.title("Mean steps per day")
plt.xlabel("Mean steps")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

Analyse if we see differences in steps for different age groups or educational background

```python
plt.figure(figsize = (15,8))
plt.xticks(rotation=45)
sns.violinplot(x='age',y='steps_mean',data=df)
plt.title('Mean steps for each age group')
plt.xlabel('Age group')
plt.ylabel('Mean steps')
plt.show()
```

It's clear that young people below 25 walk less and older people who can still walk (between 55 and 64) most, the differences within this group are however also bigger.

```python
plt.figure(figsize = (15,8))
plt.xticks(rotation=45)
sns.boxplot(x='education_1',y='steps_mean',data=df)
sns.swarmplot(x='education_1',y='steps_mean', size=3, data=df)
plt.title('Mean steps for each age group')
plt.xlabel('Age group')
plt.ylabel('Mean steps')
plt.show()
```

Educational level does not seem to influence the nr of steps that much. This graph however shows that most people have a bachelor or higher

### lineaire regressie houd in dat als variabie 1 met x toenoeemt varbaibele twee altijd met y toeneemt. Bijvoorbeeld als varibabele 1 met 1 omhoog gaat gaat varbabele twee altijd met 3 omhoog

```python
df_sub=df.loc[:,"id":"weight"]
df_sub2=df.loc[:,"steps_mean"]
df=pd.concat([df_sub,df_sub2],axis=1)
df.head()
```

```python
df_dum = pd.get_dummies(df,columns=['age','gender', 'city'])
```

```python
df_dum.head()
```

```python
df_r = df_dum.dropna()
df_r.head()
```

```python
sns.lmplot(x="steps_mean",y="weight", scatter_kws={"s":3,"alpha":0.8}, data=df_r)
```

## lm-model

```python
from sklearn.model_selection import train_test_split #this is the function we need to split the data

x = df_r.loc[:, ~df_r.columns.isin(['education_1','steps_mean','id','hh_size'])]
y = df_r['steps_mean'] 

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=80) #splits the data into train (70%) and test (30%) data
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import math
from sklearn.metrics import mean_squared_error

lm = LinearRegression() 
lm.fit(x_train, y_train)
```

```python
y_test_p = lm.predict(x_test)
r2_score(y_test, y_test_p)
```

```python
math.sqrt(mean_squared_error(y_test, y_test_p))
```

# Conclusie

#### De conclusie van mijn test is dat met de huidige waardes je niet gemakkelijk kan voorspellen hoeveel stappen er gezet worden door iemand. De r2 score is namelijk redlijk laag. Je kan met de test 2,4% van de toename in stappen voorspellen met de beschikbare gegevens.
